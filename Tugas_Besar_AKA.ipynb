```python
# Algo A : Iteratif
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def find_primes(a, b):
    primes = []
    for num in range(a, b + 1):
        if is_prime(num):
            primes.append(num)
    return primes

ranges = [(1, 25), (100, 150), (300, 450), (700, 1000)]

for a, b in ranges:
    if a < b:
        primes = find_primes(a, b)
        print(f"Iteratif [{a}, {b}]: {primes}")
```

```python
import time


# Algo B : Rekursif
def is_prime_recursive(n, i=2):

    if n < 2:
        return False
    if i > int(n**0.5):
        return True
    if n % i == 0:
        return False
    return is_prime_recursive(n, i + 1)

def find_primes_recursive(a, b, primes=None):

    if primes is None:
        primes = []
    if a > b:
        return primes
    if is_prime_recursive(a):
        primes.append(a)
    return find_primes_recursive(a + 1, b, primes)

ranges = [(1, 25), (100, 150), (300, 450), (700, 1000)]

for a, b in ranges:
    if a < b:

        # Ukur waktu eksekusi versi rekursif
        start_time = time.time()
        primes = find_primes_recursive(a, b)
        end_time = time.time()
        recursive_time = end_time - start_time
        print(f"Rekursif [{a}, {b}]: {primes}")
        print(f"Waktu eksekusi rekursif: {recursive_time:.6f} detik")
        print("-" * 20)
```

```python
import time
import matplotlib.pyplot as plt

# Fungsi untuk menentukan apakah suatu bilangan adalah prima
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Fungsi untuk menemukan bilangan prima dalam rentang tertentu (iteratif)
def find_primes(a, b):
    primes = []
    for num in range(a, b + 1):
        if is_prime(num):
            primes.append(num)
    return primes

# Fungsi untuk menemukan bilangan prima dalam rentang tertentu (rekursif)
def find_primes_recursive(a, b):
    if a > b:
        return []
    if is_prime(a):
        return [a] + find_primes_recursive(a + 1, b)
    else:
        return find_primes_recursive(a + 1, b)

# Fungsi untuk mengukur waktu eksekusi
def measure_time(func, a, b):
    start_time = time.time()
    func(a, b)
    end_time = time.time()
    return end_time - start_time

# Definisikan rentang angka yang akan diuji
ranges = [(1, 25), (100, 150), (300, 450), (700, 1000)]

# Menyimpan waktu eksekusi
iterative_times = []
recursive_times = []

# Mengukur waktu eksekusi untuk kedua algoritma
for a, b in ranges:
    # Mengukur waktu untuk algoritma iteratif
    iterative_time = measure_time(find_primes, a, b)
    iterative_times.append(iterative_time)

    # Mengukur waktu untuk algoritma rekursif
    recursive_time = measure_time(find_primes_recursive, a, b)
    recursive_times.append(recursive_time)

    print(f"Rentang [{a}, {b}] - Iteratif: {iterative_time:.5f}s, Rekursif: {recursive_time:.5f}s")

# Membuat grafik untuk perbandingan waktu
range_labels = [f"[{a}, {b}]" for a, b in ranges]

plt.plot(range_labels, iterative_times, label="Iteratif", color='b', marker='o')
plt.plot(range_labels, recursive_times, label="Rekursif", color='r', marker='x')

plt.xlabel("Rentang Angka")
plt.ylabel("Waktu Eksekusi (detik)")
plt.title("Perbandingan Waktu Eksekusi Algoritma Iteratif dan Rekursif")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45) # Rotasi label sumbu x agar lebih mudah dibaca
plt.tight_layout() # Menyesuaikan layout agar label tidak terpotong
plt.show()

# Membuat grafik untuk algoritma iteratif
range_labels = [f"[{a}, {b}]" for a, b in ranges]

plt.plot(range_labels, iterative_times, label="Iteratif", color='b', marker='o')

plt.xlabel("Rentang Angka")
plt.ylabel("Waktu Eksekusi (detik)")
plt.title("Waktu Eksekusi Algoritma Iteratif")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Membuat grafik untuk algoritma rekursif
range_labels = [f"[{a}, {b}]" for a, b in ranges]

plt.plot(range_labels, recursive_times, label="Rekursif", color='r', marker='x')

plt.xlabel("Rentang Angka")
plt.ylabel("Waktu Eksekusi (detik)")
plt.title("Waktu Eksekusi Algoritma Rekursif")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
